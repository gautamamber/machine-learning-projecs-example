##### Boston House Prediction using Linear Regression

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load data
# data the data to learn, target the regression targets
from sklearn.datasets import load_boston
boston = load_boston()
boston
```

```python
# Transform data into dataset
df_x = pd.DataFrame(boston.data, columns=boston.feature_names)
# target data
df_y = pd.DataFrame(boston.target)
```

```python
df_x
```

```python
df_y
```

```python
# get some stats from df_x or on independent variables
df_x.describe()
```

```python
reg = LinearRegression()
```

```python
# split the data
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.3, random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
# train our data
reg.fit(x_train, y_train)
```

```python
# Get the estimated coefficients for the linear regression model
# Regression coefficients are estimates of the unknown population parameters and describe the relationship between a predictor variable and the response.
print(reg.coef_)
```

```python
# Now predictions
y_pred = reg.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
# check accuracy or performance
# 1. Mean squared errorr (MSE)
# it is always non-negative, and values closer to zero indicate a better fit
np.mean((y_pred-y_test)**2)
```

```python
from sklearn.metrics import mean_squared_error
cs = mean_squared_error(y_test, y_pred)
```

```python
# The low mean square value, great predictions
cs
```
